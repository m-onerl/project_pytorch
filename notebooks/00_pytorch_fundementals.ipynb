### Getting information from tensors 

1. Get data type form tensor use tensor.dtype
2. for get shape of tensor we have to use tensor.shape
3. and when we want right debices from tensor we use tensor.device

```python
#Create a tensor

tensor = torch.rand(3, 4, dtype = torch.float16, device = 'cuda')
tensor
```

```python
#Find details about tensor
print(tensor)
```

```python
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Shape of tensor: {tensor.shape}")
print(f"Device tensor is stored on: {tensor.device}")
print(f"Size of tensor: {tensor.size}")
```
