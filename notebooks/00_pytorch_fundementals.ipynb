## Reshaping, stacking, squeezing and unsqueezing tesnors

* Reshaping - reshapes an input tensor to other shape

* View - other view but keep shape of memory as origin one

* Stacking - combine tensor to be on top of each other(vstack) or side by side using (hstack)

* Squeeze - removes 1 from tesors dimension 

* Unsquezze - add 1 to dimensions

* Permute - swapp dimension permuted in certain way

```python
x = torch.arange(1., 10.)
x , x.shape
```

```python
#Add extra dimenison

x_reshaped = x.reshape(1, 9)
x_reshaped2 = x.reshape(3, 3)
x_reshaped, x_reshaped.shape , x_reshaped2, x_reshaped2.shape   
```
